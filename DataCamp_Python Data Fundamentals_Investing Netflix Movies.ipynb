```python
# Importing pandas, matplotlib, and numpy
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
# check for NA
netflix_df.isna().sum()
```

```python
# create a new df for 1990s data
movies_90s = netflix_df[(netflix_df['release_year'] >= 1990) & (netflix_df['release_year'] < 2000)]

# preview the new df
movies_90s.head()
```

```python
# get the most frequent movie duration in the 1990s
duration = int(movies_90s['duration'].mode())
print(f'The most frequent duration among the 1990s movies is {duration} minutes.')
```

```python
# count the number of short action movies released in the 1990s

# create a new column to label the movies based on their duration
movies_90s['duration_type'] = np.where(movies_90s['duration'] < 90, 'Short', 'Long')

# preview the results
movies_90s.head()
```

```python
# count the number of 'short' action movies
short_movie_count = int(movies_90s[(movies_90s['duration_type'] == 'Short') & (movies_90s['genre'] == 'Action')]['duration_type'].value_counts())
print(f'There are {short_movie_count} short action movies in the 1990s.')
```
